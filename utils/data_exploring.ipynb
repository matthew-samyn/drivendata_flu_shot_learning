```python
for column in h1z1.columns[1:]:
    nans = h1z1[column].isna().sum()
    print(f"""
Column:  {column}
NaN's : {nans}""")
    ax = sns.catplot(kind="count", data=h1z1, x=column, hue="h1n1_vaccine")
    plt.title(f"Countplot on {column}")
    plt.ylim(0, 1500)
    plt.yticks([0,200,400,700,1000,1500])
    plt.show()
```
